```python
import matplotlib.pyplot as plt
import numpy as np
from Plot_Function import tangent_plane
from Comp_Methods import euler, euler_mod
```

```python
y0 = 1
t0 = 0
delta_t = 0.00001
```

```python
f = lambda t, y: y*(2-y)

exact_sol = lambda t: (2*y0)/(y0 + (2-y0)*np.exp(-2*t))
```

```python
euler_values = euler(f, t0, y0, 4, delta_t)[1]
euler_mod_values = euler_mod(f, t0, y0, 4, delta_t)[1]
times = np.arange(t0, 4 + delta_t, delta_t)
exact_values = exact_sol(times)
```

```python
plt.figure(figsize=(50,50))
tangent_plane(f, (0,0), (4,4), (0.1,0.1))
plt.plot(times, euler_values, label="Euler")
plt.plot(times, euler_mod_values, label="Euler Modificado")
plt.plot(times, exact_values, label="Solução Exata")
plt.axis([0, 4, 0, 4])
plt.legend()
```
