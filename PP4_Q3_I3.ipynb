```python
import matplotlib.pyplot as plt
import numpy as np
from Plot_Function import tangent_plane
from Comp_Methods import euler, euler_mod
from Convergences import total_convergence, estimated_convergence
```

```python
y0 = 1
t0 = 0
delta_t = 0.5
```

```python
f = lambda t, y: y*(2-y)

exact_sol = lambda t: (2*y0)/(y0 + (2-y0)*np.exp(-2*t))
```

```python
euler_values = euler(f, t0, y0, 4, delta_t)[1]
euler_mod_values = euler_mod(f, t0, y0, 4, delta_t)[1]
times = np.arange(t0, 4 + delta_t, delta_t)
exact_values = exact_sol(times)
```

```python
plt.figure(figsize=(10,10))
tangent_plane(f, (0,0), (4,4), (0.1,0.1))
plt.plot(times, euler_values, label="Euler", color = "black", linestyle = "dotted")
plt.plot(times, euler_mod_values, label="Euler Modificado", color = "black", linestyle = "dashed")
plt.plot(times, exact_values, label="Solução Exata", color="black")
plt.axis([0, 4, 0, 4])
plt.legend()
plt.title("Comparação de soluções")
```

```python
euler_total_conv = total_convergence(f, t0, y0, 1, 2, 10, euler, exact_sol)
euler_mod_total_conv = total_convergence(f, t0, y0, 1, 2, 10, euler_mod, exact_sol)
euler_est_conv = estimated_convergence(f, t0, y0, 1, 2, 10, euler)
euler_mod_est_conv = estimated_convergence(f, t0, y0, 1, 2, 10, euler_mod)
```

```python
def list_of_lists_to_latex_table(data):
    table = "\\begin{tabular}{|" + "c|" * len(data[0]) + "}\n"
    table += "\\hline\n"
    
    for row in data:
        table += " & ".join(map(str, row)) + " \\\\\n"
        table += "\n"
    
    table += "\\end{tabular}"
    
    return table
```
