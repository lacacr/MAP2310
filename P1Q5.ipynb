```python
import numpy as np
from P1Q5_sol import P1Q5
```

```python
w = 5
phi = lambda delta_t, f, t, y: f(y+w*delta_t*f(y))
```

```python
sol = P1Q5(phi=phi, w=w)
interval = sol._calculateAbsStabInt()
for i in range(len(interval)//2):
    print(f"Interval {i}: {-interval[2*i+1], -interval[2*i]}")
```

```python
sol._StabRegWithMonteCarlo(100000)
```

```python
amp_fac = lambda l, delta_t: abs(1 + l*delta_t + w*(l*delta_t)**2)
sol._StabRegWithLevelCurves(100, 100, amp_fac)
```
